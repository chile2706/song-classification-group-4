```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('spotify_songs.csv')
df.head()
```

```python
df['genre'], unique_numbers = pd.factorize(df['playlist_genre'])
df.head()
```

```python
unique_numbers
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
col_drop = ['track_id', 'track_name', 'track_artist', 'track_popularity', 'track_album_id', 'track_album_name', 'track_album_release_date', 'playlist_name', 'playlist_id', 'playlist_genre', 'playlist_subgenre']
df.drop(col_drop, axis=1, inplace=True)
df.head()
```

# Save data

```python
save = df.sample(frac=1).reset_index(drop=True)
save.head()
```

```python
save.shape
```

```python
save['genre'].value_counts()
```

```python
train_num_rows = int(len(save) * 0.6)
validation_num_rows = int(len(save)*0.65)
print(train_num_rows)
print(validation_num_rows)
```

```python
train_to_save = save.iloc[:train_num_rows]
valid_to_save = save.iloc[train_num_rows:validation_num_rows]
test_to_save = save.iloc[train_num_rows:]
```

```python
train_to_save.to_csv('../class_data/60_spotify_train.csv', index=False)
test_to_save.to_csv('../class_data/60_spotify_test.csv', index=False)
valid_to_save.to_csv('../class_data/60_spotify_valid.csv', index=False)
```
